```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.formula1.com/en/results/2024/drivers"
res = requests.get(url)
print(res)
```

```python
soup = BeautifulSoup(res.text, "lxml")
```

```python
F1_title = soup.select("title")
print(F1_title[0].text)
```

```python
titles = soup.select("table thead tr th")

name_titles = []
for title in titles :
    name_titles.append(title.text)
print(name_titles[1:])
```

```python
f1_lists = soup.select("table tbody tr ")

for list in f1_lists:
    col = list.find_all("td")
    # print(list.prettify())
    pos = col[0].text.strip()
    name = col[1].text.strip()
    name = name[0:-3]
    nationally = col[2].text.strip()
    car = col[3].text.strip()
    score = col[4].text.strip()
    print(f" {pos}. \n Diver: {name} \n Nationally: {nationally}\n Car: {car}\n Score: {score}\n")

```

```python
import pandas as pd
```

```python
pos_lists = []
driver_lists = []
nationally_lists = []
car_lists = []
score_lists = []
F1_listsFit = []

for list in f1_lists:
    col = list.find_all("td")

    pos_lists.append(col[0].text.strip())
    name = col[1].text.strip()
    driver_lists.append(name[0:-3])
    nationally_lists.append(col[2].text.strip())
    car_lists.append(col[3].text.strip())
    score_lists.append(col[4].text.strip())
    F1_listsFit = [driver_lists,nationally_lists,car_lists,score_lists]

dy = pd.DataFrame(F1_listsFit,index = name_titles[1:],columns = pos_lists)
dy
```
