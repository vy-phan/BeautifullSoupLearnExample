```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://coinmarketcap.com/vi/"
res = requests.get(url)
print(res)
```

```python
soup = BeautifulSoup(res.text, "lxml")
```

```python
title = soup.title.string
print(title)
```

```python
tables = soup.select("tbody tr td")
# print(tables)
for table in tables:
    name = table.find_all("p")
    for nam in name:
        print(nam.text)
```

```python
lists = soup.select("tbody tr")
for list in lists:
    name = list.find_all("td")
    for na in name[:10]:
        print(na.text)
```
